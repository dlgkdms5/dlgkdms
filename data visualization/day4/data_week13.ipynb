```python
constitution = open('/content/drive/MyDrive/Colab Notebooks/data visualization/korea.txt', encoding='UTF-8').read()
constitution
```

```python
import re
```

```python
constitution2 = re.sub('[0-9①-⑨]', '', constitution)
constitution2
```

```python
constitution3 = re.sub('\n', '', constitution2)
constitution3
```

```python
constitution3 = re.sub('\ufeff', '', constitution3)
constitution3
```

```python
nouns = hannanum.nouns(constitution3)
nouns
```

```python
import pandas as pd
df = pd.DataFrame({'word' : nouns})
df
```

```python
df['length'] = df['word'].str.len()
df
```

```python
df2 = df.query('length>=2')
df2.sort_values('length')
```

```python
df3 = df2.groupby('word', as_index=False)\
                .agg(n = ('word', 'count'))\
                .sort_values('n', ascending=False)
```

```python
df3
```

```python
top50 = df3.head(50)
top50
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 10))
sns.barplot(data=top50, x='n', y='word', hue='word')
```

```python
dic_word = top50.set_index('word').to_dict()['n']
dic_word
```
