```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rc('font', family='NanumBarunGothic')
```

```python
data = {'이름' : ['철수', '영희', '진숙', '완이'],
        '성별' : ['남', '여', '여', '남'],
        '점수' : [70,80,90,80]}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df['추가점수'] = df['점수']*0.1
df
```

```python
plt.figure(figsize=(8,4))
plt.plot(df['이름'], df['점수'], label='점수', marker='*', linestyle='-.', color='red')
plt.plot(df['이름'], df['추가점수'], label='추가점수', marker='o', linestyle='--', color='blue')
plt.title("중간고사 성적표")
plt.xlabel("이름")
plt.ylabel("점수", loc='top', rotation=0)
plt.ylim(-10,110)
plt.legend(loc=(1.02,0.7))
plt.grid()
plt.show()
```

```python
plt.pie(df['점수'],(0,0.1,0,0), df['이름'], colors=['blue', 'pink', 'green', 'red'],autopct='%1.1f%%',
        startangle=90)
plt.title("중간고사 성적표")
```

```python
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# 사용자 입력 받기
try:
    score = float(input("점수를 입력하세요 (0~100): "))
    if 0 <= score <= 100:
        grade = get_grade(score)
        print(f"당신의 학점은 {grade}입니다.")
    else:
        print("⚠️ 0에서 100 사이의 숫자를 입력하세요.")
except ValueError:
    print("⚠️ 숫자를 입력해주세요.")
```
