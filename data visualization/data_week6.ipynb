```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rc('font', family='NanumBarunGothic')
```

```python
data = {'이름' : ['철수', '영희', '진숙', '완이'],
        '성별' : ['남', '여', '여', '남'],
        '점수' : [70,80,90,80]}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df['추가점수'] = df['점수']*0.1
df
```

```python
plt.figure(figsize=(8,4))
plt.plot(df['이름'], df['점수'], label='점수', marker='*', linestyle='-.', color='red')
plt.plot(df['이름'], df['추가점수'], label='추가점수', marker='o', linestyle='--', color='blue')
plt.title("중간고사 성적표")
plt.xlabel("이름")
plt.ylabel("점수", loc='top', rotation=0)
plt.ylim(-10,110)
plt.legend(loc=(1.02,0.7))
plt.grid()
plt.show()
```

```python
plt.pie(df['점수'],(0,0.1,0,0), df['이름'], colors=['blue', 'pink', 'green', 'red'],autopct='%1.1f%%',
        startangle=90)
plt.title("중간고사 성적표")
```
