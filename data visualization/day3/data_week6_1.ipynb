```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
```

```python
import pandas as pd
df = pd.read_html('https://ko.wikipedia.org/wiki/%EA%B8%B0%EB%8C%80%EC%88%98%EB%AA%85#cite_note-:2-1')
df
```

```python
df0=df[0]
```

```python
df0
```

```python
plt.bar(df0['Unnamed: 0'], df0['2019'], color=['green', 'blue', 'red'])
plt.title("우리나라 기대수명")
plt.xlabel('성별')
plt.ylabel('수명', rotation=0)
plt.show()
```

```python
df0 = df0.rename(columns={'Unnamed: 0': '구분'})
df0
```

```python
select_c = (df0[['구분', '1970', '2000', '2016', '2019']])
select_c
```

```python
select_r = df0.iloc[1:3]
select_r
```

```python
select = df0.loc[1:3,['구분','1970','2000', '2016', '2019']]
select
```

```python
plt.plot(select['구분'], select['1970'], label='1970년')
plt.plot(select['구분'], select['2000'], label='2000년')
plt.plot(select['구분'], select['2016'], label='2016년')
plt.plot(select['구분'], select['2019'], label='2019년')
plt.title('평균수명')
plt.xlabel('성별')
newlabel=['남자수명', '여자수명']
plt.xticks(rotation=45)
plt.ylabel('수명')
plt.legend(loc=(1.04,0.7))
plt.show()
```
